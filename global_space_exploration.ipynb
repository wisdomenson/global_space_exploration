```python
#Average Mission Budget and Success Rate by Country
import matplotlib.pyplot as plt
import seaborn as sbn

sbn.set_style('dark')
#Group and prepare the data
avg_data = space_exp_ds[['Country', 'Success Rate (%)', 
                         'Budget (in Billion $)']].groupby('Country').mean().sort_values(by='Budget (in Billion $)')

#Create the plot
fig, ax1 = plt.subplots(figsize=(7, 4))
plt.xlabel('Country')

#Barchart for Budget
ax1.bar(avg_data.index, avg_data['Budget (in Billion $)'], color='skyblue', label='Avg Budget (Billion $)')
ax1.set_ylabel('Average Budget (in Billion $)', color='teal')
ax1.tick_params(axis='y', labelcolor='teal')
ax1.set_xticklabels(avg_data.index, rotation=30, ha='right')


#Create a second y-axis for Success Rate
ax2 = ax1.twinx()
ax2.plot(avg_data.index, avg_data['Success Rate (%)'], color='green', marker='o', label='Avg Success Rate (%)')
ax2.set_ylabel('Average Success Rate (%)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

#Titles and layout
plt.title('Average Mission Budget and Success Rate by Country')
fig.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xlabel('Country')
plt.show()
```

```python
#Country's Missions Impact On The Environment
import pandas as pd
import seaborn as sbn

import matplotlib.pyplot as plt

sbn.set_style('darkgrid')
# Group and count environmental impact per country
impact_counts = space_exp_ds.groupby(['Country', 'Environmental Impact']).size().unstack(fill_value=0)

# Sort countries by total missions in ascending order
impact_counts = impact_counts.loc[impact_counts.sum(axis=1).sort_values().index]

# Plot stacked bar chart
ax = impact_counts.plot(kind='bar', stacked=True, figsize=(7, 4))
plt.title("Country's Missions Impact On The Environment")
plt.xlabel('Country')
plt.ylabel('Impact Level')
plt.legend(title='Environmental Impact', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=20)
plt.ylim(top=350)
plt.tight_layout()

# Add figures to the bars
for c_idx, country in enumerate(impact_counts.index):
    cumulative = 0
    for i, impact in enumerate(impact_counts.columns):
        value = impact_counts.loc[country, impact]
        if value > 0:
            ax.text(
                c_idx, 
                cumulative + value / 2, 
                str(value), 
                ha='center', 
                va='center', 
                fontsize=8,
                color='black' 
                #color='white' if i != 0 else 'black'
            )
            cumulative += value

plt.show()
```
